#Model creation

```python
#split the data into X and Y

x = data.drop('Attrition',axis=1)
y = data.Attrition
```

```python
#training and testing data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 42)
```

```python
#balanced

data.Attrition.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE()

x_smote, y_smote = sm.fit_resample(x_train,y_train)
```

```python
#import the algorithm and train the model

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

model.fit(x_smote,y_smote)
```

Predictions -- x_test ---> are new records (unknown)

```python
y_pred = model.predict(x_test)
```

Predcitons on -- x_train ---> x_train is my actual training data

we are predictiing on training data to check the training accuracy

```python
y_pred_train = model.predict(x_train)
```

```python
#Evaluate

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)  #testing accuracy
```

```python
#traning accuracy

accuracy_score(y_train,y_pred_train) #training accuracy
```

```python
#here the training accuracy is 100, but testing accuracy is 72%

#low bias but high variance ----------> overfitting
```

To order to overcome the overfitting we need to apply hyper parameter tuning


#hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
#1 . criterion : measures the impurity of the split ---> entropy , gini

#2. splitter : best, random  it is the strategy used to decide on what basis the nodes will be splitted

#3 . max_depth :it tells the height of three

#4. min_samples_split : minimum samples required to split the decision node

#5. min_samples_leaf : minimum samples required at leaf node
```

```python
#l = [1,2,3,4,5,6,7,7,8,9,10]

print(list(range(1,20)))
```

```python
params = {
    "criterion":("gini","entropy"),
    "splitter":("best","random"),
    "max_depth":(list(range(1,20))),
    "min_samples_split":[2,3,4],
    "min_samples_leaf":list(range(1,20))
}
```

```python
#apply the Grid search CV

t = DecisionTreeClassifier(random_state = 3)
t_cv = GridSearchCV(t,params,scoring="accuracy",n_jobs=-1,verbose = 1, cv = 3)
t_cv.fit(x_smote,y_smote)
best_params = t_cv.best_params_
print(f"best params: {best_params}")
```

```python
t_cv.best_score_
```

```python
model = DecisionTreeClassifier(criterion = "entropy",max_depth = 17, min_samples_leaf = 1, min_samples_split=4,splitter = 'random')
model.fit(x_smote,y_smote)
y_pred = model.predict(x_test)
```

```python
y_pred_train = model.predict(x_smote)
```

```python
accuracy_score(y_test,y_pred) # -------testing
```

```python
accuracy_score(y_smote,y_pred_train) #training
```
