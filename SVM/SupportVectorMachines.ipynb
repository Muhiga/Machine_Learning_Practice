## Model Creation

```python
x=preprocessed_data.drop("Loan_Status",axis=1)
y=preprocessed_data["Loan_Status"]
```

```python
## creating training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x, y,test_size=0.30,random_state=3)
```

### Balancing data

```python
## balancing the data
data.Loan_Status.value_counts()
```

## Smote

```python
!pip install imblearn
```

```python
# Install imblearn package - pip install imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE()
```

```python
# Smote is always applied on train data(x_train and y_train)
x_smote, y_smote = smote.fit_resample(x_train,y_train)
```

**Counter** is a container which keeps track to how many times equivalent values are added. Python counter class is a part of collections module and is a subclass of dictionary.

```python
from collections import Counter
print("Actual Classes",Counter(y_train))
print("SMOTE Classes",Counter(y_smote))
```

 Support Vector Classifier Model

```python
from sklearn.svm import SVC
svclassifier = SVC() ## base model with default parameters # if regressor its SVR
svclassifier.fit(x_smote, y_smote) #x_train , y_train -> after smote i.e balancing the data
```

Predict output for X_test

```python
y_pred=svclassifier.predict(x_test)
```


evaluating the model created

Accuracy


```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,classification_report,f1_score
acc=accuracy_score(y_test,y_pred)
acc
```

f1 score

```python
f1_score(y_test,y_pred)
```

Calssification report

```python
# Calssification report measures the quality of predictions. True Positives, False Positives, True negatives and False Negatives
#are used to predict the metrics of a classification report

print(classification_report(y_test,y_pred))
```

cross validation score

```python
## checking cross validation score
from sklearn.model_selection import cross_val_score

scores=cross_val_score(svclassifier,x,y,cv=10,scoring='f1')
print(scores)

print("cross validation score",scores.mean())

print("standard deviation",scores.std())

# std <0.05 , then model is good otherwise bad model.
```

```python
# gridsearch cv

from sklearn.model_selection import GridSearchCV

# define hyperparameter range   #parameter grid
#paramter key:vlaue  c:[], g:[],
param_grid={'C':[0.1,5,10,50,60,70],'gamma':[1,0.1,0.01,0.001,0.001],'random_state':list(range(1,20))}

model=SVC()

grid=GridSearchCV(model,param_grid,verbose=2,scoring="f1",cv=5)
#cv= 5 that is entire data is dibvided intp 5 diff datasets i.e 5 diff training and testing dta
#on these 5 diff sets
#all the parameter c and gamma is applied along with the model SVC
#ans the performance is evaluated.
#verbose = 2 --> just to print the sentences, prints msgs,statement


grid.fit(x,y) # activation of grid search CV

```
